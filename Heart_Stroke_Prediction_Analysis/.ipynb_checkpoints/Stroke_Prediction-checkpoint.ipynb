```python
number_of_women = data[data['gender'] == 'Female'].shape[0]
number_of_men = data[data['gender'] == 'Male'].shape[0]
number_women_with_stroke = data[(data['gender'] == 'Female') & (data['stroke'] == 1)].shape[0]
number_men_with_stroke = data[(data['gender'] == 'Male') & (data['stroke'] == 1)].shape[0]
number_women_without_stroke = data[(data['gender'] == 'Female') & (data['stroke'] == 0)].shape[0]
number_men_without_stroke = data[(data['gender'] == 'Male') & (data['stroke'] == 0)].shape[0]
```

```python
Percentage_men_with_stroke =  float(number_men_with_stroke/number_of_men)
Percentage_women_with_stroke = float(number_women_with_stroke/number_of_women)
Percentage_men_without_stroke = float(number_men_without_stroke/number_of_men)
Percentage_women_without_stroke = float(number_women_without_stroke/number_of_women)
```

```python

print('Percentage of men with chance of stroke ',Percentage_men_with_stroke*100)
print('Percentage of women with chance of stroke ',Percentage_women_with_stroke*100)
print('Percentage of men without chance of stroke ',Percentage_men_without_stroke*100)
print('Percentage of women without chance of stroke ',Percentage_women_without_stroke*100)
```

**The percentage of men with chance of stroke is very close to that of percentage of women with the chances of stroke**

```python
data['work_type'].unique()
```

```python
Private = data[data['work_type'] == 'Private']
Self_Employed = data[data['work_type'] == 'Self-employed']
Govt_Job = data[data['work_type'] == 'Govt_Job']
Home_Maker = data[data['work_type'] == 'children']
Never_worked  = data[data['work_type'] == 'Never_worked']
```

```python
 l =[Private,Self_Employed,Govt_Job,Home_Maker,Never_worked]
```

```python
occupational_stroke = list()
```

```python
def compute_stroke_pecent(df):
    if df['stroke'].value_counts(normalize=True).shape[0] == 2:
        occupational_stroke.append(df['stroke'].value_counts(normalize=True)[1])
    else:
        occupational_stroke.append(0)
    
```

```python
for df in l:
    compute_stroke_pecent(df)
    
```

```python
print(occupational_stroke)
```

```python
from prettytable import PrettyTable
x = PrettyTable(['work_type','Chance of stroke in percentage'])
Resultant = dict(zip(list(data['work_type'].unique()),occupational_stroke))
print(Resultant)
```

```python
for key,value in Resultant.items():
    
    row = []
    row.append(key)
    row.append(np.round(value,3))
    x.add_row(row)

print(x)
```

**From the above we see that chances of stroke among Self-employed is higher, follwed by those in Private and then Homemakers**

```python
data.smoking_status.value_counts()
```

```python
never_smoked = data[data['smoking_status'] == 'never smoked']
unknown = data[data['smoking_status'] == 'Unknown']
formerly_smoked = data[data['smoking_status'] == 'formerly smoked']
smokes = data[data['smoking_status'] == 'smokes']


l =[never_smoked,unknown,formerly_smoked,smokes]

smoke_stroke = list()

def compute_stroke_pecent(df):
    if df['stroke'].value_counts(normalize=True).shape[0] == 2:
        smoke_stroke.append(df['stroke'].value_counts(normalize=True)[1])
    else:
        smoke_stroke.append(0)
    


for df in l:
    compute_stroke_pecent(df)
    

print(smoke_stroke)

from prettytable import PrettyTable
x = PrettyTable(['smoker_type','Chance of stroke in percentage'])
Resultant = dict(zip(list(data['smoking_status'].unique()),smoke_stroke))
print(Resultant)


for key,value in Resultant.items():
    
    row = []
    row.append(key)
    row.append(np.round(value,3))
    x.add_row(row)

print(x)
```

**It is seen that for people who smoke the chance of stroke is the highest , followed by those who have smoked previously**
