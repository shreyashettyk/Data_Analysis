# Central Limit Theorm

```python
black_friday_sales = pd.read_csv('https://raw.githubusercontent.com/mathubhalan/Black-Friday-Sales/master/Data/BlackFriday.csv')
```

```python
black_friday_sales.head()
```

```python
black_friday_sales.isnull().sum()
```

```python
data = black_friday_sales['Purchase']
```

```python
data.mean()
```

```python
sample_means = list()
def compute_mean_from_samples(data,number_of_samples,sample_size):
    sample_means = list()
    for i in range(number_of_samples):
        samples = random.sample(range(0,data.shape[0]),sample_size)
        df = data.loc[samples]
        sample_mean = df.mean()
        sample_means.append(sample_mean)
    
    return sample_means
```

```python
#check if sample distributions of sample menas follow normal distribution or not
def check_for_normal(means_from_samples):
    stats.probplot(means_from_samples,dist='norm',plot=pylab)
```

```python
def central_limit_theorm(mean_values_from_samples,key):
    population_mean = np.mean(mean_values_from_samples)
    print('population mean with ',i,'is',population_mean)
```

```python
sample_central_limit = dict()
sample_central_limit['sample_10_size_50'] = compute_mean_from_samples(data,10,50)
sample_central_limit['sample_100_size_50'] = compute_mean_from_samples(data,100,50)
sample_central_limit['sample_1000_size_50'] = compute_mean_from_samples(data,1000,50)
sample_central_limit['sample_10000_size_50'] = compute_mean_from_samples(data,10000,50)
sample_central_limit['sample_100000_size_50'] = compute_mean_from_samples(data,100000,50)
```

```python
for i in sample_central_limit.keys():
    central_limit_theorm(sample_central_limit[i],i)
```

```python
measurements = sample_central_limit['sample_10_size_50']
stats.probplot(measurements,dist='norm',plot=pylab)
```

**Observations**

For the above CLT experimentation we have taken varied number of samples all of size 50
We find that as the number of samples increases, they align more with the normal distribution

With more number of samples, the consecutive diffrence between the two population means is also observed to have reduced
